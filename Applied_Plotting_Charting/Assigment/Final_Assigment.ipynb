```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('Exportaciones_Fruta.csv')
df2 = pd.read_csv('dolar.csv')
df1 = df1.iloc[:,0:7]
df2 = df2.iloc[:,:2].set_index('Fecha')
df2.index = pd.to_datetime(df2.index)
```

```python
years = np.arange(df2.index.min().year,df2.index.max().year,1)
avg_year = []
for year in years:
    d=[]
    for i in range(len(df2)):
        if df2.index[i].year == year:
            d.append(df2['Valor Dólar'][i])
    avg_year.append(np.mean(d))
```

```python
df1['Income $MM'] = df1['Exportaciones US$FOB']/10**9
```

```python
%matplotlib notebook
fig = plt.figure()

ax1 = plt.subplot(211)
plt.plot(df1['Año'],df1['Income $MM'],'-o')
plt.ylabel('Dolars Income [$ MM]')
plt.title('Trend Comparison \n Dolars Income by Fruit Exportation and Dolar Value in Chile')

ax2 = plt.subplot(212,sharex=ax1)
plt.plot(years,avg_year,'-o')
plt.xlabel('Year')
plt.ylabel('Dolars Value')
```
